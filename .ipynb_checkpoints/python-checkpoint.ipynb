```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("Spotify_Song_Attributes_Cleaned.csv")
```

```python
genre_col = "major_genre"
value_col = "valence"

plot_df = df[[genre_col, value_col]].dropna()
plot_df = plot_df[~plot_df[genre_col].isin(["Unknown", "Other"])]

genre_order = (
    plot_df.groupby(genre_col)[value_col]
    .median()
    .sort_values()
    .index
    .tolist()
)

fig = px.box(
    plot_df,
    x=genre_col,
    y=value_col,
    color=genre_col,
    category_orders={genre_col: genre_order},
    color_discrete_sequence=px.colors.qualitative.Dark24,
    points="outliers",
    title="Valence by Genre"
)

fig.update_layout(
    xaxis_title="Genre",
    yaxis_title="Valence (Musical Positivity)",
    template="plotly_white",
    width=1100,
    height=650,
    legend_title_text="Genre"
)

fig.show()
```
